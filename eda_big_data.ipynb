**6.3.4 Observation:**
 The `M` and `k` are the unique values in the `Size` column, that represents Mbs and Kbs. Now we will convert them into one common unit of bytes, and then change this column's data type from `object` to `numeric`.

### 6.3.4 Finding Count values

```python
# Find the count of values in the 'Size' column that contain 'M'
m_count = df['Size'].loc[df['Size'].str.contains('M')].value_counts().sum()

# Find the count of values in the 'Size' column that contain 'k'
k_count = df['Size'].loc[df['Size'].str.contains('k')].value_counts().sum()

# Print the results
print(f"Count of values in the 'Size' column that contain 'M': {m_count}")
print(f"Count of values in the 'Size' column that contain 'k': {k_count}")
```

### 6.2.5 Define a Function

Let's convert the `M` and `K` units into bytes and then remove the `M` and `K` from the values and convert them into numeric data type.

```python
# convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
# this function will convert the size column to numeric
def convert_size(size):
    # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
    '''
    
    if isinstance(size, str):
        size = size.replace(',', '')  # Remove commas from the string
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
    return size

df['Size'] = df['Size'].apply(convert_size)
```

### 6.3.6 Convert Size column into numeric data tpye

```python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)

# Convert 'Size_in_bytes' column to numeric
df['Size_in_bytes'] = pd.to_numeric(df['Size_in_bytes'], errors='coerce')

# Create a new column 'Size_in_Mb' with size in MB
df['Size_in_Mb'] = df['Size_in_bytes'] / (1024 * 1024)

# making a new column called 'Size in Mb' which will have the size in MB
df['Size_in_Mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

* Now we have converted every value into bytes and removed the `M` and `K` from the values and converted them into numeric data type.

### 6.3.8 Imputing missing values of 'Installs' column

```python
# Remove non-numeric characters from 'Installs' column
df['Installs'] = df['Installs'].str.replace('[^\d]', '', regex=True)

# Convert 'Installs' column to numeric type
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Calculate the median of 'Installs' column
median_installs = df['Installs'].median()

# Fill missing values in 'Installs' column with the median
df['Installs'].fillna(median_installs, inplace=True)
```

### 6.3.9 Removing signs of `+` and `,` from 'Installs' column

```python
# remove the plus sign from install column and convert it to numeric
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
# also remove the commas from the install column
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
# convert the install column to numeric (integers because this is the number of installs/count)
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

#### Q.15: What are the most common pricing ranges for paid apps?

```python
import pandas as pd

# Assuming your DataFrame is named df
# df = pd.read_csv('your_file.csv')

# Filter the DataFrame to include only paid apps
paid_apps = df[df['Price'] > 0]

# Define the pricing ranges
pricing_ranges = {
    '0-1': (0, 1),
    '1-5': (1, 5),
    '5-10': (5, 10),
    '10-20': (10, 20),
    '20+': (20, float('inf'))
}

# Create a new column 'Price Range' based on the pricing ranges
def get_price_range(price):
    for range_name, (lower, upper) in pricing_ranges.items():
        if lower <= price < upper:
            return range_name
    return 'Unknown'

paid_apps['Price Range'] = paid_apps['Price'].apply(get_price_range)

# Count the occurrences of each pricing range
common_pricing_ranges = paid_apps['Price Range'].value_counts()

# Display the most common pricing ranges for paid apps
print(common_pricing_ranges)
```
